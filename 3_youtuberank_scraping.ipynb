```python
from selenium import webdriver
```

```python
browser=webdriver.Chrome('./chromedriver.exe')
```

```python
browser.get('https://youtube-rank.com/board/bbs/board.php?bo_table=youtube')
```

```python
html=browser.page_source
```

```python
from bs4 import BeautifulSoup
```

```python
soup=BeautifulSoup(html, 'html.parser')
```

```python
contents=soup.select('tr.aos-init')
print(type(contents))
contents[0]
```

```python
content= contents[0]
```

```python
content.select('p.category')[0]
```

```python
category=content.select('p.category')[0]
```

> title.text.strip()

* first=title.text
* second=first.strip()


```python
for content in contents:
    category=content.select('p.category')[0]
    title=content.select('h1>a[href*="board"]')[0]
    print(category.text.strip(),title.text.strip())
```

```python
songs=[]#or list()
for content in contents:
    category=content.select('p.category')[0]
    title=content.select('h1>a[href*="board"]')[0]
    print(category.text.strip(),title.text.strip())
    songs.append([category.text.strip(),title.text.strip()])
songs
```

```python
import pandas as pd
```

```python
pd_data=pd.DataFrame(songs)
pd_data
```

```python
pd_data=pd.DataFrame(songs, columns=['제목','가수'])
pd_data
```
