```python
from selenium import webdriver
```

```python
browser=webdriver.Chrome('./chromedriver.exe')
```

```python
browser.get('http://ticket.interpark.com/Contents/Ranking?smid1=s_menu&smid2=performance')
```

```python
html=browser.page_source
```

```python
from bs4 import BeautifulSoup
```

```python
soup=BeautifulSoup(html,'html.parser')
len(soup),type(soup)
```

```python
soup.select('div.items')
```

```python
tags=soup.select('div.items')
len(tags),type(tags)
```

```python
tags[0]
```

```python
tag=tags[0]
```

```python
name=tag.select('a.prdName')
len(name),type(name)
```

```python
name[0].text.strip()
```

```python
date=tag.select('a.prdDuration')
len(date),type(date)
```

```python
date[0].text.strip()
```

```python
for tag in tags:
    name=tag.select('a.prdName')
    date=tag.select('a.prdDuration')
    print(name[0].text.strip(),date[0].text.strip())
```

```python
contents=list()
for tag in tags:
    name=tag.select('a.prdName')
    date=tag.select('a.prdDuration')
    musical=[name[0].text.strip(),date[0].text.strip()]
    contents.append(musical)
contents
```

```python
import pandas as pd
```

```python
pd.DataFrame(contents)
```

```python
pd_data=pd.DataFrame(contents, columns=['Title','Date Range'])
pd_data
```

```python
pd_data.to_excel('./saves/interpark_scraping.xls', index=False)
```
