```python
from selenium import webdriver
```

```python
browser=webdriver.Chrome('./chromedriver.exe')
```

```python
browser.get('https://www.jobkorea.co.kr/')
```

```python
browser.find_elements_by_css_selector('li>a.linkItem')[0].click()
```

```python
from bs4 import BeautifulSoup as bs
```

```python
html=bs(browser.page_source, 'html.parser')
```

```python
tags=html.select('ul.recommend-lists--common>li.list-item')
tags
```

```python
tag=tags[0]
tag
```

회사명:a.corp-name

직군:p.content

위치:p.local

```python
corp=tag.select('a.corp-name')[0]
content=tag.select('p.content')[0]
loc=tag.select('p.local')[0]
```

```python
job=[]
for i in tags:
    corp=i.select('a.corp-name')[0]
    content=i.select('p.content')[0]
    loc=i.select('p.local')[0]
   # print(corp.text.strip(),content.text.strip(),loc.text.strip())
    job.append([corp.text.strip(),content.text.strip(),loc.text.strip()])
job
    
```

```python
import pandas as pd
```

```python
pd_data=pd.DataFrame(job, columns=['회사명','직무','위치'])
pd_data
```

```python
pd_data.to_excel('./saves/4_jobkorea_ws0629.xls',index=False)
```
