```python
browser.get('https://www.genie.co.kr/chart/top200')
```

```python
from bs4 import BeautifulSoup
```

```python
len(soup),type(soup)
```

```python
for tag in tags:
    title=tag.select('a.title')
    artist=tag.select('a.artist')
    print(title[0].text.strip(),artist[0].text.strip())
```

-------------- 판다스에 넣기 -------------------
* list만들어내기

[[title1,artist1],[title2,artist2],[title3,artist3],..]

```python
contents=list()
for tag in tags:
    title=tag.select('a.title')
    artist=tag.select('a.artist')
#     print(title[0].text.strip(),artist[0].text.strip())
    song=[title[0].text.strip(),artist[0].text.strip()]
    contents.append(song)

contents
```

-----------데이터 프레임으로 만들어 저장------------

```python
import pandas as pd
```

```python
pd.DataFrame(contents)
```

```python
pd_data=pd.DataFrame(contents, columns=['title','artist'])
pd_data
```

```python
pd_data.to_excel('./saves/gennie_scraping.xls', index=False)
```
